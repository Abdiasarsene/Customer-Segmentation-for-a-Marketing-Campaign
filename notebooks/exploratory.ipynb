## Importing librairies

```python
# Importing librairies
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import missingno as msno
```

## Importing dataset

```python
# Importing dataset
customer = pd.read_csv(r"../data/customer_segmentation.csv")
print("Jeu de dçnnées importé ✅✅")
```

## Printing dataset with Missingno

```python
msno.bar(customer, color='red')
```

## Separation of data's types

```python
num_col = customer.select_dtypes(include=["int64","float64"]).columns.tolist()
cat_col = customer.select_dtypes(include=['object']).columns.tolist()
```

## Distribution of Numeric Variables

```python
customer[num_col].hist(bins=30, figsize=(15,8))
plt.suptitle("Distribution of numercial variables", fontsize=16)
plt.show()
```

## Heat Map

```python
plt.figure(figsize=(10,8))
sns.heatmap(customer[num_col].corr(), annot=True, cmap="coolwarm")
plt.title("Correlation between numeric variables")
plt.show()
```

## Distribution of Categorical Variables

```python
for col in cat_col:
    sns.countplot(data=customer, x=col, order=customer[col].value_counts().index)
    plt.title(f"Distribution of {col}")
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()
```

```python
sns.countplot(customer, x=customer["Preferred_Category"], hue=customer["Used_Discount"])
plt.title("Preferred_Category vs Used_Discount")
plt.show()
```

## Data display

```python
sns.pairplot(customer, hue="Used_Discount", diag_kind="kde",corner=True)
plt.show()
```
